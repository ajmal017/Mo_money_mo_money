## EDA on TSLA 

It would be interesting to see if there are any general trends within TSLA stock.
Why TSLA? It's an exploding stock that's becoming less and less correlated with company valuation, and more running on hype. Analysis on this could lead to measuring objective values of people's hype psychology (FOMO, panic, etc.)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta, time, datetime


import warnings
from IPython.core.display import display, HTML
warnings.filterwarnings("ignore") # suppress warnings because theyre annoying and i like living on the edge
pd.set_option('display.max_columns',100) # so i can see all columns
np.set_printoptions(linewidth = 220) # increase numpy array print
display(HTML("<style>.container { width:80% !important; }</style>")) # widen jupyter notebook blocks
```

```python
work_dir = 'D:/5_us_txt/data/5 min/us/nasdaq stocks/2/'
df = pd.read_csv(work_dir + 'tsla.us.txt')
```

```python
cols = ['ticker','period','date','time',
        'open','high','low','close','vol','openint']
df.columns = cols
```

```python
# datetime in PST
df['datetime'] = pd.to_datetime(df['date'].astype(str) + df['time'].astype(str), format = '%Y%m%d%H%M%S') - timedelta(hours = 9)

# time in PST
df['time'] = (df['time'] / 100).astype(int) - 900 
```

```python
df.head()
```

```python
## check to see correlation between volume and open/close and high/low diff price

df['diff_open_close'] = df['close'] - df['open'] # in dollars, negative means stock price went down
df['diff_max_min'] = df['high'] - df['low'] # in dollars, must be non-negative
df['mid'] = (df['high'] + df['low']) / 2. # average of high and low

```

```python
## see the diff of open/close over time
cond = df['date'] == 20200706
t = df[cond]['time']
y1 = df[cond]['diff_open_close']
y2 = df[cond]['diff_max_min']
plt.figure(figsize = [10,5])
plt.plot([np.min(t), np.max(t)], [0,0])
plt.scatter(t,y1)
plt.xlabel('time')
plt.ylabel('dollars')
plt.title('plot time against diff in open/close')
plt.show()

plt.figure(figsize = [10,5])
plt.scatter(t,y2)
plt.xlabel('time')
plt.ylabel('dollars')
plt.title('plot time against diff in max/min')
plt.show()

```

```python
## see the diff of open/close over volume 


cond = df['date'] == 20200706
plt_df = df[cond].copy()

# flat price

x = plt_df['vol']
y1 = plt_df['diff_open_close']
y2 = plt_df['diff_max_min']
plt.figure(figsize = [10,5])
plt.scatter(x,y1)
plt.plot([np.min(x), np.max(x)], [0,0])
plt.xlabel('volume')
plt.ylabel('dollars')
plt.title('plot volume against diff in open/close')
plt.show()

plt.figure(figsize = [10,5])
plt.scatter(x,y2)
plt.xlabel('volume')
plt.ylabel('dollars')
plt.title('plot volume against diff in max/min')
plt.show()

# proportional (relative) price to open price

x = plt_df['vol']
y1 = plt_df['diff_open_close'] / plt_df['open'] * 100
y2 = plt_df['diff_max_min'] / plt_df['open'] * 100
plt.figure(figsize = [10,5])
plt.scatter(x,y1)
plt.plot([np.min(x), np.max(x)], [0,0])
plt.xlabel('volume')
plt.ylabel('relative dollar (%)')
plt.title('plot volume against relative diff in open/close')
plt.show()

plt.figure(figsize = [10,5])
plt.scatter(x,y2)
plt.xlabel('volume')
plt.ylabel('relative dollar (%)')
plt.title('plot volume against relative diff in max/min')
plt.show()



```
