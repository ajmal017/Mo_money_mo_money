```python
ticker = yf.Ticker("AAPL")
data = ticker.history(period = '1d',
                   interval = '5m').reset_index()
data['Datetime'] = data.Datetime.map(lambda x: x.replace(tzinfo=timezone('US/Pacific')))
```

```python
data['time'] = data['Datetime'].map(lambda x: x.time)
```
