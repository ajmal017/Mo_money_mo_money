```python
df = msft.history(period="max")
df.reset_index(inplace = True)
```

```python
from datetime import timedelta
div_date = df[df.Dividends != 0]['Date'].tolist()
div_date_after = (df[df.Dividends != 0]['Date'] + timedelta(days=1)).tolist()
div_date_before = (df[df.Dividends != 0]['Date'] + timedelta(days=-1)).tolist()

all_dates = div_date + div_date_after + div_date_before
```
